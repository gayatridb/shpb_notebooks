```python
import pandas as pd
data = pd.read_excel('shpb.xlsx', index_col=0)
```

```python
data
```

```python
ndata = data[["Time","Voltage_0", "Voltage_1"]].copy()
```

```python
print (ndata)
```

```python
f = 9.48 * 10**(-4)
# calculated voltage = c
c = 0.75;
# Incident strain = ei
# Transmission strain = et
#ei = 100 * (vi * f)/(2 * c);
#et = 100 * (vt * f)/(2 * c);
ndata['ei'] = ndata.apply(lambda row: 100 * (row['Voltage_0'] * f)/(2 * c), axis=1)
ndata['et'] = ndata.apply(lambda row: 100 * (row['Voltage_1'] * f)/(2 * c), axis=1)
```
